```python
import os
import pandas as pd
```

```python
BASE_DIR = os.path.dirname(os.getcwd())
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
os.makedirs(CACHE_DIR, exist_ok=True)
```

```python
csv_files = [files for files in os.listdir(DATA_DIR) if files.endswith('.csv')]
csv_files
```

```python
filename = '2005.csv'
filepath = os.path.join(DATA_DIR, filename)
temp_df = pd.read_csv(filepath)
temp_df['filename'] = filename
year = filename.replace(".csv", '')
temp_df['year'] = year
temp_df.head(n=10)
```

```python
all_dataframes = []

for filename in csv_files:
    csv_path = os.path.join(DATA_DIR, filename)
    temp_df = pd.read_csv(csv_path)
    temp_df['filename'] = filename
    year = filename.replace(".csv", '')
    temp_df['year'] = year
#     temp_df.head(n=10)
    all_dataframes.append(temp_df)
```

```python
combined_movie_dataset = pd.concat(all_dataframes)
```

```python
dataset_path = os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
combined_movie_dataset.to_csv(dataset_path, index=False)
```
